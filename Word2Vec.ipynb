**NLP Lecture 5 (19-12-19):** Using Word2Vec to train model for similarity scores

```
import pandas as pd
import numpy as np
import gensim
import gzip
import nltk
from gensim.models import Word2Vec
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```
data=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
```

```
nltk.download('punkt')
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
```

```
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~''' #Function to remove punctuations
def punct_remover(str):
  no_punct = ""
  for char in str:
    if char not in punctuations:
      no_punct = no_punct + char
  return no_punct
```

```
def stopwords_remover(text): #Tokenizing text and removing the stop words
  text=text.lower()
  tokens=word_tokenize(text)
  nostop=[words for words in tokens if not words in stop_words]
  return nostop
```

```
data['reviewText']=data['reviewText'].apply(punct_remover)
```

```
data['revtxt_nostop_tokens']=data['reviewText'].apply(stopwords_remover)
```

```
def unique_tokens(text):
  unique_list=[]
  for words in text:
    if words not in unique_list:
      unique_list.append(words)
  return unique_list
```

```
data['revtxt_unique_tokens']=data['revtxt_nostop_tokens'].apply(unique_tokens)
```

```
data.head()
```

```
wordlist=[[word] for word in data['revtxt_nostop_tokens'][0]]
```

```
wordlist
```

```
model=Word2Vec(wordlist, size=100, window=5, min_count=1, workers=4)
```

```
for i in wordlist:
  for j in wordlist:
    if j>i:
      print(i[0],' :',j[0],model.similarity(i[0],j[0]))
    else:
      continue
```
